```python
from causationentropy import discover_network
from causationentropy.datasets.synthetic import linear_stochastic_gaussian_process
from causationentropy.graph import network_to_dataframe



data, A = linear_stochastic_gaussian_process(rho=0.7, n=5, T=200, p=0.5, seed=42)
network = discover_network(data, method='standard', max_lag=5)
df = network_to_dataframe(network)
df.head()
```

```python
from causationentropy.core.plotting import plot_causal_network

fig, ax = plot_causal_network(network,
    save_path="network.png", 
    label_fontsize = 48,
    title_fontsize = 48,
    colorbar_fontsize = 30,
    legend_fontsize = 30,
    stats_fontsize = 25       
)
```
